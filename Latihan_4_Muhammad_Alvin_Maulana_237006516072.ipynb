```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from skimage import data
```

```python
img = data.coins().astype(np.float32)

plt.figure(figsize=(5,5))
plt.imshow(img, cmap='gray')
plt.title("Original Coins")
plt.axis('off')
```

```python
sigma = 15
noise = np.random.normal(0, sigma, img.shape)
noisy_img = img + noise
noisy_img = np.clip(noisy_img, 0, 255)

plt.figure(figsize=(5,5))
plt.imshow(noisy_img, cmap='gray')
plt.title("Coins + Gaussian Noise ( = 15)")
plt.axis('off')
```

```python
median_filtered = cv2.medianBlur(noisy_img.astype(np.uint8), ksize=5)
gaussian_filtered = cv2.GaussianBlur(noisy_img, (5,5), sigmaX=1)
bilateral_filtered = cv2.bilateralFilter(noisy_img.astype(np.uint8), d=9, sigmaColor=75, sigmaSpace=75)

plt.figure(figsize=(14,5))

plt.subplot(1,4,1)
plt.imshow(noisy_img, cmap='gray')
plt.title("Noisy Image")
plt.axis('off')

plt.subplot(1,4,2)
plt.imshow(median_filtered, cmap='gray')
plt.title("Median Filter")
plt.axis('off')

plt.subplot(1,4,3)
plt.imshow(gaussian_filtered, cmap='gray')
plt.title("Gaussian Filter")
plt.axis('off')

plt.subplot(1,4,4)
plt.imshow(bilateral_filtered, cmap='gray')
plt.title("Bilateral Filter")
plt.axis('off')

plt.show()
```

```python
edges_noisy = cv2.Canny(noisy_img.astype(np.uint8), 100, 200)
edges_median = cv2.Canny(median_filtered, 100, 200)
edges_gaussian = cv2.Canny(gaussian_filtered.astype(np.uint8), 100, 200)
edges_bilateral = cv2.Canny(bilateral_filtered, 100, 200)

plt.figure(figsize=(14,6))

plt.subplot(2,2,1)
plt.imshow(edges_median, cmap='gray')
plt.title("Canny after Median Filter")
plt.axis('off')

plt.subplot(2,2,2)
plt.imshow(edges_gaussian, cmap='gray')
plt.title("Canny after Gaussian Filter")
plt.axis('off')

plt.subplot(2,2,3)
plt.imshow(edges_bilateral, cmap='gray')
plt.title("Canny after Bilateral Filter")
plt.axis('off')

plt.subplot(2,2,4)
plt.imshow(edges_noisy, cmap='gray')
plt.title("Canny on Noisy Image (Baseline)")
plt.axis('off')

plt.show()
```

```python
print("""
=== ANALISIS HAASIL LATIHAN 4 ===

1. Median Filter
    Paling efektif menghilangkan noise impuls dan Gaussian ringan.
    Namun dapat membuat tepi sedikit tumpul.
    Peta tepi Canny menjadi lebih bersih, tapi detail kecil hilang.

2. Gaussian Filter
    Menghaluskan citra secara merata.
    Mengurangi noise dengan baik tetapi mengaburkan tepi.
    Edge map menunjukkan garis tepi lebih tipis dan halus.

3. Bilateral Filter
    Terbaik dalam menjaga tepi (edge-preserving).
    Menghilangkan noise tanpa menghilangkan detail tepi penting.
    Edge map paling akurat dan bersih dibanding metode lain.

4. Kesimpulan:
    Median: bagus untuk noise, kurang bagus untuk detail tepi.
    Gaussian: halus, tapi tepi semakin blur.
    Bilateral: hasil terbaiktepi terjaga, noise turun signifikan.
""")
```
